```python
import kagglehub

path = kagglehub.dataset_download(
    "abubakerasiel/global-data-science-and-analytics-job-dataset"
)

print("Dataset path:", path)
```

```python
import os

os.listdir(path)
```

```python
import pandas as pd
import os

file_path = os.path.join(path, "job_postings_monthly.xlsx")

df = pd.read_excel(file_path)

df.head()
```

```python
from ast import literal_eval

skills = df['job_skills'].dropna().apply(literal_eval)

all_skills = []
for row in skills:
    all_skills.extend(row)

import pandas as pd
pd.Series(all_skills).value_counts().head(10)
```
